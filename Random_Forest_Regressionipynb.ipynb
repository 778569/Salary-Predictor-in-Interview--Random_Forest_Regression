```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
X = dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=20, random_state=3)
#in here passing number of trees
model=rf.fit(X,y)
```

```python
model.predict([[6.5]])
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y,color='red')
plt.plot(X,model.predict(X),color='green')

plt.title('RFR for Salaries Predictor')
plt.xlabel('Positions')
plt.ylabel('Salaries')
plt.show()
```

```python
X_grid = np.arange(min(X),max(X),0.01)
print(X_grid)
```

```python
X_grid = X_grid.reshape((len(X_grid),1))
print(X_grid)
```

```python
pln_X_grid = model.predict(X_grid)
plt.scatter(X,y,color='red')
plt.plot(X_grid, model.predict(X_grid),color='green')

plt.title('RFR for Salaries Predictor')
plt.xlabel('Positions')
plt.ylabel('Salaries')
plt.show()
```
